¬øQu√© es el operador l√≥gico NOT?

NOT es el siguiente paso l√≥gico y adem√°s introduce algo clave: el peso negativo.

El operador NOT hace lo contrario de la entrada, para que NOT funcione necesitamos:

* que cuando x = 0 ‚Üí salida = 1
* que cuando x = 1 ‚Üí salida = 0

CLAVE: eso se logra con un peso negativo.

La neurona hace: suma = w ‚ãÖ x  y luego eval√∫a: suma >= umbral

Par√°metros correctos para NOT

Usaremos:
* Entrada: una sola variable
* Peso: -1
* Umbral: 0
¬øPor qu√©?

Si x = 0 ‚Üí suma = 0 ‚Üí se activa
Si x = 1 ‚Üí suma = -1 ‚Üí no se activa


```python
#Importamos las librer√≠as que ocuparemos
#Se usa NumPy para manejar arreglos y hacer el producto punto (dot product) entre dos arreglos.

import numpy as np
# Entradas para la neurona (una sola variable)

X = np.array([
    [0],
    [1]
])

# Salidas para NOT
Y = np.array([1, 0])
# Peso negativo. Aqu√≠ est√° el operador NOT, aunque no aparece con la palabra.
W = [-1]
# Umbral para la funci√≥n de activaci√≥n.
umbral = 0
```

```python
#Constructor
class McCullochPitts:
    def __init__(self, weights, umbral): # Guarda los pesos y el umbral dentro del objeto.
    # Se calcula w1‚Äãx1‚Äã+w2‚Äãx2‚Äã
        """
            Constructor de la neurona MCP.
            Guarda las variables
            weights -> vector de pesos iniciales
            umbral -> umbral para usar en la funcion de activaci√≥n
        """
        self.weights = weights
        self.umbral = umbral

# M√©todo predict
    def predict(self, X):
        """
            Calcula la salida de la neurona para cada fila de entradas X.
        """
        # Lista donde se guardar√°n las salidas
        predictions = []
        # Recorremos cada fila de entradas
        for i in range(X.shape[0]):

            # Calculamos la suma ponderada: w1*x1 + w2*x2
            weighted_sum = np.dot(self.weights, X[i])

            # Mostramos en consola la entrada y la suma ponderada
            print(f"Input: {X[i]}, Weighted Sum: {weighted_sum}")

            # Comparaci√≥n con el umbral determinar si la neurona se activa o no

            if weighted_sum >= self.umbral:
                predictions.append(1)
            else:
                predictions.append(0)
        return predictions
```

```python
# Instanciamos la clase McCullochPitts
mcp = McCullochPitts(W, umbral)
# Calculamos las predicciones
predictions = mcp.predict(X)
# Mostramos el resultado final
print("Predictions:", predictions)
```

En este ejemplo, con un umbral igual a 1, la neurona se activa cuando al menos una de las entradas es 1, reproduciendo correctamente el comportamiento del operador l√≥gico OR
